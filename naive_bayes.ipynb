```python
BNoDeliveryNo = 4/9
BYesDeliveryNo = 5/9

BNoDeliveryYes = 10/41
BYesDeliveryYes = 31/41

BNoDelivery = 14/50
BYesDelivery = 36/50
```

```python
# Kondisi B dimana
# Day = Holiday
# Discount = Yes
# Free Delivery = Yes
# P(A) = NoBuy = BNo
# P(B) = BYesDiscount, BYesDelivery, BHoliday
# P(B|A)1 = (BHoliday|No Buy) = BHolidayNo
# P(B|A)2 = (BYesDiscount|No Buy) = BYesDiscountNo
# P(B|A)3 = (BYesDelivery|No Buy) = BYesDeliveryNo
# P(B)1 = BHoliday
# P(B)2 = BYesDiscount
# P(B)3 = BYesDelivery

# Ditanyakan P(A|B) ?
# Rumus
# P(A|B) = P(B|A) * P(A) / P(B)
```

```python
BNoBuy1 = ((BHolidayNo*BYesDiscountNo*BYesDeliveryNo)*BNo)/(BHoliday*ByesDiscount*BYesDelivery)
print(BNoBuy1)
```

```python
# Kondisi B dimana
# Day = Holiday
# Discount = Yes
# Free Delivery = Yes
# P(A) = Buy = BYes
# P(B) = BYesDiscount, BYesDelivery, BHoliday
# P(B|A)1 = (BHoliday|Buy) = BHolidayYes
# P(B|A)2 = (BYesDiscount|Buy) = BYesDiscountYes
# P(B|A)3 = (BYesDelivery|Buy) = BYesDeliveryYes
# P(B)1 = BHoliday
# P(B)2 = BYesDiscount
# P(B)3 = BYesDelivery

# Ditanyakan P(A|B) ?
# Rumus
# P(A|B) = P(B|A) * P(A) / P(B)
```

```python
BBuy2 = ((BHolidayYes*BYesDiscountYes*BYesDeliveryYes)*BYes)/(BHoliday*ByesDiscount*BYesDelivery)
print(BBuy2)
```

```python
# probability of No buy = 0.12989982125784594
# Probability of buy = 0.8440667832398943
```

```python
BSum = BNoBuy1+BBuy2
print(BSum)
```

```python
BBuy = (BBuy2/BSum) *100
print(BBuy)
```

```python
BNoBuy = (BNoBuy1*BSum) *100
print(BNoBuy)
```
